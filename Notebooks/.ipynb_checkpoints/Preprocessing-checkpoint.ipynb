# Missing Value

```python
# Create a table showing the number of missing data points and the percentage.
miss_count = df.isnull().sum()
miss_percent = miss_count / len(df) * 100
miss_df = pd.DataFrame({'Missing Count': miss_count, 'Missing Percent': miss_percent}).sort_values('Missing Percent', ascending = False)
miss_df
```

```python
#Remove columns where more than 50 percent of the data is missing.
thresh = 50

to_drop = miss_df[miss_df['Missing Percent'] > 50].index.to_list()
df = df.drop(columns = to_drop)
```

```python
# Split the data into numbers and object
num_cols = df.select_dtypes(include = [np.number]).columns.to_list()
cat_cols = df.select_dtypes(include = ['object','category','bool']).columns.to_list()
```

```python
# Flag creation is missing
cols_with_flag = df.columns[df.isnull().sum() > 0]
for col in cols_with_flag:
    df[f"{col}_was_missing"] = df[col].isnull().astype(int)

df.head()
```

```python
#Fill in the data using the median by group.
if "Neighborhood" in df.columns:
    df[num_cols] = df.groupby('Neighborhood')[num_cols].transform(lambda x: x.fillna(x.median()))
```

```python
# Fill in the data numeric
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
```

```python
# Fill category using the mode
for col in cat_cols:
    if df[col].isnull().any():
        mode = df[col].mode()
        fill_na = mode.iloc[0] if len(mode) else "Missing"
        df[col] = df[col].fillna(fill_na)
```

```python
# Ordinal features with limited unique values
ordinal_condition = nunique[(nunique >= 10) & (nunique <= 20)].index
ordinal_feature = [col for col in ordinal_condition if col not in zero_feature and col not in flag and col not in year_feature]
ordinal = ordinal_feature
ordinal
```

```python
# Plot histogram and boxplot
def plot_hist_box(df, col):
    data = df[col]
    
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    plt.figure(figsize=(10, 4))

    # Histogram
    plt.subplot(1, 2, 1)
    sns.histplot(data, kde = True, bins=30)
    plt.axvline(lower_bound, linestyle="--", label="Lower bound")
    plt.axvline(upper_bound, linestyle="--", label="Upper bound")
    plt.title(f"Histogram of {col}")
    plt.legend()

    # Boxplot 
    plt.subplot(1, 2, 2)
    sns.boxplot(data, vert=False)
    plt.axvline(lower_bound, linestyle="--")
    plt.axvline(upper_bound, linestyle="--")
    plt.title(f"Boxplot of {col}")

    plt.tight_layout()
    plt.show()
```

## Preprocessing Outlier Report

```python
# Create Summary Continuouse
continuous_report = []
for col in continuous:
    series = df[col]
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = ((series < lower)|(series > upper))
    outlier_count = outliers.sum()
    outlier_percent = outlier_count / len(series) * 100

    # XÃ¡c Ä‘á»‹nh hÃ nh Ä‘á»™ng Ä‘á» xuáº¥t
    skew_val = series.skew()
    if abs(skew_val) > 1:
        action = "log transform"
    elif outlier_percent < 5:
        action = "capping preferred over median imputation"
    elif outlier_percent < 15:
        action = "winsorize"
    else:
        action = "review distribution"

    continuous_report.append({
        "Column": col,
        "Feature Type": "continuous",   # hoáº·c zero_inflated / ordinal / count / year
        "Q1": Q1,
        "Q3": Q3,
        "Lower Bound": lower,
        "Upper Bound": upper,
        "Outlier Count": outlier_count,
        "Outlier %": outlier_percent,
        "Recommended Action": action
    })

continuous_report_df = pd.DataFrame(continuous_report)
continuous_report_df
```

```python
# Create Summary Zero_inflated
zero_inflated_report = []
for col in zero_inflated:
    action = "split zero and non_zero  / Winsorize"
    serius = df[col]
    Q1 = serius.quantile(0.25)
    Q3 = serius.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = ((series < lower)|(series > upper))
    outlier_count = outliers.sum()
    outlier_percent = outlier_count / len(series) * 100
    zero_inflated_report.append({
        "Column": col,
        "Feature Type": "zero_inflated",   # zero_inflated / ordinal / count / year
        "Q1": Q1,
        "Q3": Q3,
        "Lower Bound": lower,
        "Upper Bound": upper,
        "Outlier Count": outlier_count,
        "Outlier %": outlier_percent,
        "Recommended Action": action
    })

zero_inflated_report_df = pd.DataFrame(zero_inflated_report)
zero_inflated_report_df
```

```python
# Create Summary Ordinal
ordinal_report = []
for col in ordinal:
    action = "Treat as ordinal categorical â€” no IQR logic"
    serius = df[col]
    Q1 = serius.quantile(0.25)
    Q3 = serius.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = ((series < lower)|(series > upper))
    outlier_count = outliers.sum()
    outlier_percent = outlier_count / len(series) * 100
    ordinal_report.append({
        "Column": col,
        "Feature Type": "Ordinal", 
        "Q1": Q1,
        "Q3": Q3,
        "Lower Bound": lower,
        "Upper Bound": upper,
        "Outlier Count": outlier_count,
        "Outlier %": outlier_percent,
        "Recommended Action": action
    })

ordinal_report_df = pd.DataFrame(ordinal_report)
ordinal_report_df
```

```python
# Create Summary Count
count_report = []
for col in count:
    action = "No outliers handling"
    serius = df[col]
    Q1 = serius.quantile(0.25)
    Q3 = serius.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = ((series < lower)|(series > upper))
    outlier_count = outliers.sum()
    outlier_percent = outlier_count / len(series) * 100
    count_report.append({
        "Column": col,
        "Feature Type": "Count", 
        "Q1": Q1,
        "Q3": Q3,
        "Lower Bound": lower,
        "Upper Bound": upper,
        "Outlier Count": outlier_count,
        "Outlier %": outlier_percent,
        "Recommended Action": action
    })

count_report_df = pd.DataFrame(count_report)
count_report_df
```

```python
# Create Summary Year/Time
time_report = []
for col in year_cols:
    action = "No IQR, Check Logic"
    serius = df[col]
    Q1 = serius.quantile(0.25)
    Q3 = serius.quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = ((series < lower)|(series > upper))
    outlier_count = outliers.sum()
    outlier_percent = outlier_count / len(series) * 100
    time_report.append({
        "Column": col,
        "Feature Type": "Year/ Time", 
        "Q1": Q1,
        "Q3": Q3,
        "Lower Bound": lower,
        "Upper Bound": upper,
        "Outlier Count": outlier_count,
        "Outlier %": outlier_percent,
        "Recommended Action": action
    })

time_report_df = pd.DataFrame(time_report)
time_report_df
```

```python
# Use winsor for skew > 2 and log for skew > 1
def winsorize(data, lower_q = 0.01, upper_q = 0.99):
    lower = data.quantile(lower_q)
    upper = data.quantile(upper_q)
    return data.clip(lower, upper)

skew_report = []
SPECIAL_NO_LOG = ["TotalBsmtSF"]
for col in continuous:
    raw = df[col].astype(float)

    # Skew Before
    skew_before = raw.skew()

    # Copy and print: col, skew
    transformed = raw.copy()
    print(f"{col}: {skew_before}")

    #winsorize with transformed
    if abs(skew_before) > 2:
        transformed = winsorize(transformed)

    skew_after = transformed.skew()

    # log with transformed
    if col not in SPECIAL_NO_LOG and abs(skew_after) > 1:
        transformed = np.log1p(transformed)

    # Create the log columns
    skew_after = transformed.skew()
    df[col + "_log"] = transformed

    skew_report.append({
        "Columns":col,
        "Feature Type": "Continuous",
        "Skew Before": skew_before,
        "Skew After": skew_after
    })

report_skew_df = pd.DataFrame(skew_report)
report_skew_df
```

```python
plot_hist_box(df,'LotArea_log')
```

```python
plot_hist_box(df, 'GrLivArea_log')
```

## Zero-inflated

```python
skew_nonzero_report = []
for col in zero_inflated:
    df[col] = df[col].astype(float)
    zero_mask = df[col] == 0
    
    non_zero = df.loc[~zero_mask, col]

    # Skew Before
    skew_before = non_zero.skew()

    # winsorize non-zero only
    non_zero_win = winsorize(non_zero)
    df.loc[~zero_mask, col] = non_zero_win

    # Skew After
    skew_after = non_zero_win.skew()

    # add flags with value
    df[col + "_nonzero"] = (~zero_mask).astype(int)

    skew_nonzero_report.append({
        "Columns": col,
        "Feature Type": "Non Zero",
        "Skew Before": skew_before,
        "Skew After": skew_after
    })
    
skew_nonzero_report_df = pd.DataFrame(skew_nonzero_report)
skew_nonzero_report_df
```

# Feature Engineering

```python
def add_feature_engineer(df):
    df["HouseAge"] = df["YrSold"] - df["YearBuilt"] # Tuá»•i nhÃ  
    df["YearsSinceRemod"] = df["YrSold"] - df["YearRemodAdd"] # Tuá»•i pháº§n sá»­a chá»¯a
    df["Remodeled"] = (df["YearBuilt"] != df["YearRemodAdd"]).astype(int) # NhÃ  tá»«ng sá»­a chá»¯a
    df["GarageAge"] = (df["YrSold"] - df["GarageYrBlt"]).clip(lower = 0).fillna(0)  # Garage Age
    df["TotalArea"] = (df["GrLivArea"] + df["TotalBsmtSF"] + df["GarageArea"] + df["1stFlrSF"] + df["2ndFlrSF"]) # Tá»•ng diá»‡n tÃ­ch sá»­ dá»¥ng
    df["TotalPorchSF"] = (df["OpenPorchSF"] + df["EnclosedPorch"] + df["3SsnPorch"]+ df["ScreenPorch"]) # Tá»•ng diá»‡n tÃ­ch hiÃªn nhÃ  
    df["TotalFinishedSF"] = df["1stFlrSF"] + df["2ndFlrSF"] + df["TotalBsmtSF"] # Tá»•ng diá»‡n tÃ­ch sÃ n hoÃ n thiá»‡n
    df["TotalBath"] = (df["FullBath"] + df["HalfBath"]*0.5 + df["BsmtFullBath"] + df["BsmtHalfBath"]*0.5) # Tá»•ng sá»‘ phÃ²ng táº¯m
    df["BedroomRatio"] = df["BedroomAbvGr"] / df["TotRmsAbvGrd"] # Tá»‰ lá»‡ phÃ²ng ngá»§ / Tá»•ng diá»‡n tÃ­ch
    df["KitchenRatio"] = df["KitchenAbvGr"] / df["TotRmsAbvGrd"] # Tá»‰ lá»‡ báº¿p
    df["HasBasement"] = (df["TotalBsmtSF"] > 0).astype(int) # CÃ³ basement hoÃ n thiá»‡n
    df["HasGarage"] = (df["GarageArea"] > 0).astype(int) # cÃ³ garage hay khÃ´ng 
    df["HasFireplace"] = (df["Fireplaces"] > 0).astype(int) # cÃ³ fireplace hay khÃ´ng
    df["OverallGrade"] = df["OverallQual"] + df["OverallCond"] # Tá»•ng cháº¥t lÆ°á»£ng nhÃ 
    df["HasLotFrontage"] = (~df["LotFrontage"].isna()).astype(int) # Äáº¥t cÃ³ fontage hay khÃ´ng
    df["LogGrLivArea"] = np.log1p(df["GrLivArea"]) # Log diá»‡n tÃ­ch chÃ­nh
    df["LogTotalArea"] = np.log1p(df["TotalArea"]) # Log tá»•ng diá»‡n tÃ­ch 
    df["LogLotArea"] = np.log1p(df["LotArea"]) # Log lotArea

    return df
```

```python
df_encode = add_feature_engineer(df_encode)
```
